# Detalhe de Vendas - Montando Visualização de Validação

```python
fat_dv = detalhe_de_vendas[['Codigo_Asesor','Nombre_Asesor','Codigo_Vendedor','Nombre_Vendedor','Codigo_Zona_Ruta','Venta_SO_ML_sin_IVA']].groupby(['Codigo_Asesor','Nombre_Asesor','Codigo_Vendedor','Nombre_Vendedor','Codigo_Zona_Ruta']).sum()
display(fat_dv)
```

# Detalhe de Vendas - Renomeando colunas

```python
fat_dv = fat_dv.rename(columns={'Venta_SO_ML_sin_IVA': 'Venta_SO_ML_sin_IVA_DV'})
```

# Incentivos Familia - Montando Visualização de Validação

```python
fat_if = incentivosfamilia[['Codigo_Asesor','Nombre_Asesor','Codigo_Vendedor','Nombre_Vendedor','Codigo Zona-Ruta','Venta_SO_ML_sin_IVA']].groupby(['Codigo_Asesor','Nombre_Asesor','Codigo_Vendedor','Nombre_Vendedor','Codigo Zona-Ruta']).sum()
display(fat_if)
```

# Incentivos Familia - Renomeando colunas

```python
fat_if = fat_if.rename(columns={'Venta_SO_ML_sin_IVA': 'Venta_SO_ML_sin_IVA_IF', 'Codigo Zona-Ruta' : 'Codigo_Zona_Ruta'})
```

# JOIN entre os dois reportes

```python
df_merged = pd.concat([fat_dv, fat_if], axis=1, ignore_index=True, sort=False)
df_merged = df_merged.rename(columns={0: 'Venta_SO_ML_sin_IVA_DV', 1 : 'Venta_SO_ML_sin_IVA_IF'})
teste = df_merged['Venta_SO_ML_sin_IVA_DV'] - df_merged['Venta_SO_ML_sin_IVA_IF']
df_merged = pd.concat([df_merged, teste], axis=1, ignore_index=True, sort=False)
df_merged = df_merged.rename(columns={0: 'Venta_SO_ML_sin_IVA_DV', 1 : 'Venta_SO_ML_sin_IVA_IF', 2 : 'Diff'})

#df_merged.info()
df_merged = df_merged.sort_values(by=['Diff'], ascending=False)
df_merged_vendedor = df_merged.loc[(df_merged['Diff'] >= nStart) | (df_merged['Diff'] <= nStart2)]

display(df_merged)
#display(df_merged_vendedor)
```
