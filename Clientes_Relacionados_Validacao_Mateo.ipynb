```python
# 1ª CELULA

import os
import pandas as pd
from pathlib import Path
import numpy as np


# 2ª CELULA
separator = ";"
archivo = "MC1_INT_CSV_Clientes_Relacionados_2024_01_24_05_30_22.CSV"
path = Path("./")
path.mkdir(parents=True, exist_ok=True)

print('Pasta criada com sucesso!')

path_clientes_relacionados_2024_01_24_05_30_22 = r"./clientesrelacionados/" + archivo
clientesrelacionados_2024_01_24_05_30_22 = pd.read_csv(path_clientes_relacionados_2024_01_24_05_30_22, sep=separator)


clientesrelacionados_df =  pd.DataFrame({
    'Codigo_PDV': clientesrelacionados_2024_01_24_05_30_22['Codigo_PDV']
    ,'Codigo_Tipo_de_Negocio_PDV': clientesrelacionados_2024_01_24_05_30_22['Codigo_Tipo_de_Negocio_PDV']
    ,'Descripcion_Tipo_de_Negocio_PDV': clientesrelacionados_2024_01_24_05_30_22['Descripcion_Tipo_de_Negocio_PDV']
    ,'Codigo_Cluster_PDV': clientesrelacionados_2024_01_24_05_30_22['Codigo_Cluster_PDV']
    ,'Descripcion_Cluster_PDV': clientesrelacionados_2024_01_24_05_30_22['Descripcion_Cluster_PDV']
    })


clientes_relacionados_vacios = clientesrelacionados_df[
    
     clientesrelacionados_df.Codigo_Tipo_de_Negocio_PDV.isna() 
    | clientesrelacionados_df.Descripcion_Tipo_de_Negocio_PDV.isna()
    
    | clientesrelacionados_df.Codigo_Cluster_PDV.isna()
    #| clientesrelacionados_2024_01_24_05_30_22.Descripcion_Cluster_PDV.isna()
]   


clientes_relacionados_vacios = clientes_relacionados_vacios.set_index('Codigo_PDV')

type(clientes_relacionados_vacios)

display(clientes_relacionados_vacios)

```
